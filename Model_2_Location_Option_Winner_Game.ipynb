```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Step 1: Load the dataset
df = pd.read_csv("/content/M2_Location_Option_Winner.csv", names=['venue', 'decision_team_won'])  # Add header names

# Step 2: Clean data
df.dropna(inplace=True)
df['decision_team_won'] = df['decision_team_won'].str.lower().str.strip()

# Step 3: Group by venue and decision type
venue_decision_counts = df.groupby(['venue', 'decision_team_won']).size().unstack(fill_value=0)

# Step 4: Calculate probabilities
venue_decision_probs = venue_decision_counts.div(venue_decision_counts.sum(axis=1), axis=0)

# Step 5: Display as table
print("\nWin Probability by Venue and Decision:")
print(venue_decision_probs)

# Step 6: Plot
plt.figure(figsize=(12, 6))
venue_decision_probs.plot(kind='bar', stacked=True, colormap='coolwarm', figsize=(14, 6))
plt.title("Win Probability by Venue: Bat First vs Field First")
plt.xlabel("Venue")
plt.ylabel("Winning Probability")
plt.xticks(rotation=45, ha='right')
plt.legend(title='Decision that led to win')
plt.tight_layout()
plt.grid(True)
plt.show()
```
